LIBRARY MANAGEMENT SYSTEM

```python
  
class Library:

  def __init__(self):
    # books.txt dosyasını açmak için kullanılan constructor metodu
    self.file_name= "book.txt"
    self.file=open(self.file_name,"a+")

  def __del__(self):
        # Dosyayı kapatmak için kullanılan destructor metodu
        self.file.close()

  def list_books(self):
      # books.txt dosyasındaki tüm kitapları listeleme metodu
      self.file.seek(0)  # Dosya imleci başa taşınır.Dosyanın başından itibaren okuma yapabilmek için
      lines = self.file.read().splitlines() #splitlines: her satırı bir öğe olarak içeren bir liste oluşturur.
      for line in lines:
          #Her satırı virgül karakterine göre böler ve elde edilen parçaları book_info adlı bir liste içine yerleştirir.
          book_info = line.split(',') 
          book_title, book_author, release_date, num_pages = book_info
          print(f"Kitap: {book_title}, Yazar: {book_author}")

  def add_book(self):
        # books.txt dosyasına yeni bir kitap ekleme metodu
        book_title = input("Kitap adını girin: ")
        book_author = input("Yazarı girin: ")
        release_date = input("Yayın tarihini girin: ")
        num_pages = input("Sayfa sayısını girin: ")

        book_info = f"{book_title},{book_author},{release_date},{num_pages}\n"

        # Dosyaya yazma işlemi
        self.file.write(book_info)

        self.file.flush()  # Veriyi hemen dosyaya yazmak için buffer'ı temizler
        
        print("Kitap başarıyla eklendi!")
        
  def remove_book(self):
        # books.txt dosyasından belirli bir kitabı silme metodu
        book_title_to_remove = input("Silmek istediğiniz kitabın adını girin: ")

        self.file.seek(0)
        lines = self.file.read().splitlines()
        books = [line.split(',') for line in lines]

        for i, book in enumerate(books):
            # Kitabın listedeki konumunu bulur ve listeden siler
            if book[0] == book_title_to_remove:
                del books[i]
                break
            
        # Dosyanın içeriğini tamamen siler        
        self.file.truncate(0)
        self.file.seek(0)
        
        # Güncellenmiş kitap listesini dosyaya yazarak, dosyanın içeriğini günceller
        for book in books:
            self.file.write(','.join(book) + '\n')
            
        self.file.flush()  # Veriyi hemen dosyaya yazmak için buffer'ı temizler

        print("Kitap başarıyla silindi!")
```

```python
# "Library" sınıfından bir nesne olan "lib" oluştur
lib = Library()

# "lib" nesnesiyle etkileşimde bulunmak için bir menü oluştur
while True:
    print("*** MENÜ ***")
    print("1) Kitapları Listele")
    print("2) Kitap Ekle")
    print("3) Kitap Sil")
    print("4) Çıkış")

    choice = input("Seçiminizi yapın (1-4): ")

    if choice == '1':
    # Kitapları listeleme işlemi
        lib.list_books()
    elif choice == '2':
        # Kitap ekleme işlemi
        lib.add_book()
        
        # Kullanıcıya çıkış yapmak isteyip istemediğini sormak icin
        exit_choice = input("Çıkış yapmak istiyor musunuz? (evet/hayır): ")
        if exit_choice.lower() == 'evet':
            print("Programdan çıkılıyor. İyi günler!")
            break
    elif choice == '3':
        # Kitap silme işlemi
        lib.remove_book()
        
        exit_choice = input("Çıkış yapmak istiyor musunuz? (evet/hayır): ")
        if exit_choice.lower() == 'evet':
            print("Programdan çıkılıyor. İyi günler!")
            break
    elif choice == '4':
        # Programdan çıkış işlemi
        print("Programdan çıkılıyor. İyi günler!")
        break
    else:
        # Geçersiz bir seçim yapıldığında hata mesajı göstermek icin
        print("Geçersiz seçim. Lütfen 1 ile 4 arasında bir sayı girin.")
```
